```python
import pandas as pd

df = pd.read_csv('ship_fuel_efficiency.csv')

df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['fuel_per_nm'] = df['fuel_consumption'] / df['distance']
df['co2_per_nm'] = df['CO2_emissions'] / df['distance']

df[['ship_id','distance','fuel_consumption','fuel_per_nm','co2_per_nm']]
```

```python
ship_dimension = df[['ship_id','ship_type','fuel_type']].drop_duplicates().reset_index(drop=True)
ship_dimension
```

```python
route_dimension = df[['route_id']].drop_duplicates().reset_index(drop=True)
route_dimension
```

```python
df.describe()
```

```python
def fuel_grade(fuel):
    if fuel < 21.1:
        return 'A'
    elif fuel < 34.2:
        return 'B'
    else:
        return 'C'

df['eff_grade'] = df['fuel_per_nm'].apply(fuel_grade)
```

```python
def co2_grade(co2):
    if co2 < 58.6:
        return 'A'
    elif co2 < 95.0:
        return 'B'
    else:
        return 'C'

df['co2_grade'] = df['co2_per_nm'].apply(co2_grade)
```

```python
df.groupby('fuel_type')[['fuel_per_nm','co2_per_nm']].mean()
```

```python
fuel_fact = df[['ship_id','route_id','month','distance','fuel_consumption','CO2_emissions','weather_conditions','engine_efficiency','fuel_per_nm','co2_per_nm','eff_grade','co2_grade']]

fuel_fact
```

```python
import os
import urllib.parse
from sqlalchemy import create_engine
from dotenv import load_dotenv

load_dotenv()

db_user = os.getenv("DB_USER")
db_password = os.getenv("DB_PASSWORD")
db_host = os.getenv("DB_HOST")
db_port = os.getenv("DB_PORT")
db_name = os.getenv("DB_NAME")

if db_port is None:
    db_user = "postgres"
    db_password = ""
    db_host = "localhost"
    db_port = "5432"
    db_name = "postgres"


engine = create_engine(f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')

try:
    ship_dimension.to_sql('ship_dimension', engine, if_exists='replace', index=False)
    print("1. ship_dimension 적재 완료!")

    route_dimension.to_sql('route_dimension', engine, if_exists='replace', index=False)
    print("2. route_dimension 적재 완료!")

    fuel_fact.to_sql('fuel_fact', engine, if_exists='replace', index=False)
    print("3. fuel_fact 적재 완료! 최종 ETL 성공!")

except Exception as e:
    print("적재 중 에러 발생:", e)
```
