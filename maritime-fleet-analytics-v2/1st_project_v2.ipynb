```python
df['fuel_per_nm'] = df['fuel_consumption'] / df['distance']
df['co2_per_nm'] = df['CO2_emissions'] / df['distance']

df[['ship_id','distance','fuel_consumption','fuel_per_nm','co2_per_nm']]
```

```python
ship_dimension = df[['ship_id','ship_type','fuel_type']].drop_duplicates().reset_index(drop=True)
ship_dimension
```

```python
route_dimension = df[['route_id']].drop_duplicates().reset_index(drop=True)
route_dimension
```

```python
df.describe()
```

```python
def fuel_grade(fuel):
    if fuel < 21.1:
        return 'A'
    elif fuel < 34.2:
        return 'B'
    else:
        return 'C'

df['eff_grade'] = df['fuel_per_nm'].apply(fuel_grade)
```

```python
def co2_grade(co2):
    if co2 < 58.6:
        return 'A'
    elif co2 < 95.0:
        return 'B'
    else:
        return 'C'

df['co2_grade'] = df['co2_per_nm'].apply(co2_grade)
```

```python
df.groupby('fuel_type')[['fuel_per_nm','co2_per_nm']].mean()
```

```python
month_map = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
             'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}

month_data = [{'month':m} for m in df['month'].unique()]
month_dimension = pd.DataFrame(month_data)
month_dimension['month_num'] = month_dimension['month'].map(month_map)
month_dimension['month_id'] = range(1, len(month_dimension) + 1)

fuel_fact = df[['ship_id','route_id','month','distance','fuel_consumption','CO2_emissions','weather_conditions','engine_efficiency','fuel_per_nm','co2_per_nm','eff_grade','co2_grade']]
fuel_fact = fuel_fact.merge(month_dimension[['month','month_id']], on='month', how='left')
fuel_fact = fuel_fact.drop(columns=['month'])

fuel_fact
```
